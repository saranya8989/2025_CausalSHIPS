```python
import torch.nn as nn
import torch
import torch.functional as F
# %% Define simple MLP model with 4 hidden layers, ReLU activation in hidden layers, and tanh activation in output layer
class MLP(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, hidden_size)
        self.fc3 = nn.Linear(hidden_size, hidden_size)
        self.fc4 = nn.Linear(hidden_size, hidden_size)
        self.output = nn.Linear(hidden_size, output_size)
        self.relu = nn.ReLU()
        self.tanh = nn.Tanh()
 
    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.fc3(x)
        x = self.relu(x)
        x = self.fc4(x)
        x = self.tanh(x)
        x = self.output(x)
        return x
```

```python
# Read configuration file
config_set = read_config.read_config('./config.ini')
seeds = np.arange(0,7,1)
target='DELV24'
store_nofs = miss.read_pickle('/work/FAC/FGSE/IDYST/tbeucler/default/saranya/causal/ships_plus_new/results/4/shipsold/results_fold_6.pkl')
store_corr = miss.read_pickle('/work/FAC/FGSE/IDYST/tbeucler/default/saranya/causal/ships_plus_new/results/4/shipsold/results_fold_1.pkl')

Xnorml = store_nofs['dataframes']

def get_range(Xnorml,lagindx,target,TYPE):
    """
    Xnorml: Normalized X data in the stored dictionary used to train ML model
    lagindx: 24 Hour LT (4); 48 Hour LT (8) etc.
    target: DELV24/DELV48 etc.
    TYPE: train/valid/test
    """
    ranges = {}
    current_index = 0
    
    for key in Xnorml[TYPE].keys():
        n = len(Xnorml[TYPE][key].dropna()[target][lagindx:])
        ranges[key] = (current_index, current_index + n)
        current_index += n
    return ranges

def pearson_corr(measureds, estimations):
    # Calculate the Pearson correlation coefficient
    corr = pearsonr(measureds, estimations)
    return corr.statistic

def get_stormdict_noFS(score_nocausal,foldnum, TYPE='R2'):
    """
    Computes stormwise R2 or PCC for NO feature selection results.
    """
    # Get prediction and truth for this fold
    best_pred_test = score_nocausal[foldnum]['scoreboard']['test']['pred']
    y = score_nocausal[foldnum]['y']

    PCC = []
    stormlen = []
    ypred = []
    ytruth = []

    for teststorm in causal_range_test.keys():
        storminfo = causal_range_test[teststorm]
        if storminfo[1] - storminfo[0] < 2:
            PCC.append(np.nan)
            stormlen.append(0)
            ypred.append(np.nan)
            ytruth.append(np.nan)
        else:
            stormpred = best_pred_test[storminfo[0]:storminfo[1]]
            stormy = y['test'][storminfo[0]:storminfo[1]]
            if TYPE == 'PCC':
                stormPCC = pearson_corr(stormy, stormpred)**2
            elif TYPE == 'R2':
                stormPCC = r2_score(stormy, stormpred)
            PCC.append(stormPCC)
            stormlen.append(len(stormy))
            ypred.append(stormpred)
            ytruth.append(stormy)

    stormdict = {
        'name': list(causal_range_test.keys()),
        'PCC': PCC,
        'length': stormlen
    }

    timeseries = {
        'name': list(causal_range_test.keys()),
        'ypred': ypred,
        'ytruth': ytruth
    }

    return stormdict, timeseries
```

```python
def calc_pcc_to_plot_figure5(seeds,target,config_set,folds,exp,scoretype):
    score_nocausal = []
    for seed in tqdm(seeds):
        nocausal_results = performance_scores.scores_seeds(seed=seed,target=target,lag=int(config_set['target_lag']),exp=exp).run_score_noFS()
        score_nocausal.append(nocausal_results)
        del nocausal_results
        gc.collect()

    pcc_train_noFS = [score_nocausal[i]['scoreboard']['train']['r2_from_pcc'] for i in range(len(score_nocausal))]
    pcc_valid_noFS = [score_nocausal[i]['scoreboard']['valid']['r2_from_pcc'] for i in range(len(score_nocausal))]
    pcc_test_noFS = [score_nocausal[i]['scoreboard']['test']['r2_from_pcc'] for i in range(len(score_nocausal))]

    r2_train_noFS = [score_nocausal[i]['scoreboard']['train']['r2'] for i in range(len(score_nocausal))]
    r2_valid_noFS = [score_nocausal[i]['scoreboard']['valid']['r2'] for i in range(len(score_nocausal))]
    r2_test_noFS = [score_nocausal[i]['scoreboard']['test']['r2'] for i in range(len(score_nocausal))]

    stormdict_noFS = [get_stormdict_noFS(score_nocausal,foldz, scoretype)[0] for foldz in folds]
    timeseries_noFS = [get_stormdict_noFS(score_nocausal,foldz, scoretype)[1] for foldz in folds]
    return stormdict_noFS,timeseries_noFS,score_nocausal
```

```python
device='cpu'
X_proc = score_nocausal_plus[izlc]['X']['train']

def get_pdp(Xtrain=score_nocausal_plus[izlc]['X']['train'],feat_idx=None,model=None):
    xs = Xtrain.copy()[:, feat_idx]
    q = np.quantile(xs, [0.01, 0.99])
    grid = np.linspace(q[0], q[1], 50)
    pd_vals = []
    with torch.no_grad():
        for v in grid:
            Xv = Xtrain.copy()
            Xv[:, feat_idx] = v
            t = torch.FloatTensor(Xv).to(device)
            y = model.eval()(t).squeeze(-1).cpu().numpy()
            pd_vals.append(y)
    return pd_vals,grid

def get_pdp_mlr(Xtrain=score_nocausal_plus[izlc]['X']['train'],feat_idx=None,model=None):
    xs = Xtrain.copy()[:, feat_idx]
    q = np.quantile(xs, [0.01, 0.99])
    grid = np.linspace(q[0], q[1], 50)
    pd_vals = []
    for v in grid:
        Xv = Xtrain.copy()
        Xv[:, feat_idx] = v
        y = np.squeeze(model.predict(Xv))
        pd_vals.append(y)
    return pd_vals,grid
```

```python
feat_idx = [var_names_plus.index('POT'),var_names_plus.index('T200')]
feat_idx_mlr = [var_names_plus_mlr.index('POT'),var_names_plus_mlr.index('T200')]
PDPs_SHIPSplus = [get_pdp(feat_idx=obj,model=model) for obj in feat_idx]
PDPs_SHIPSplus_MLR = [get_pdp_mlr(Xtrain=X_mlr,feat_idx=obj,model=mlr_model) for obj in feat_idx_mlr]
```
