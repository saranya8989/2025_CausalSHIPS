```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import random
import matplotlib.lines as mlines

# === Step 1: Load pickle ===
with open(f'box_plot_pkl/pccbox_shipsplus_{target}.pkl', 'rb') as f:
    pccbox = pickle.load(f)

print(" Loaded keys:", pccbox.keys())

# === Step 2: Inspect structure (do NOT convert to np.array directly) ===
for k, v in pccbox.items():
    print(f"{k}: folds={len(v)}, fold0 type={type(v[0])}, fold0 value={v[0]}")

# === Helper functions ===
def flatten_all(folds_list):
    flat_vals = []
    for fold in folds_list:
        if isinstance(fold, (list, np.ndarray)):
            flat_vals.extend(fold)
        else:
            flat_vals.append(fold)
    return np.array(flat_vals)

def fold_stats(folds_list):
    stats = []
    for fold in folds_list:
        vals = np.array(fold) if isinstance(fold, (list, np.ndarray)) else np.array([fold])
        q1, median, q3 = np.percentile(vals, [25, 50, 75])
        mn, mx = np.min(vals), np.max(vals)
        stats.append({'min': mn, 'q1': q1, 'median': median, 'q3': q3, 'max': mx})
    return stats

# === Step 3: Prepare data for plotting ===
methods = ['noFS', 'causalFS', 'corrFS', 'XAIFS']
splits = ['train', 'valid', 'test']

box_data = []
fold_stats_all = []

for split in splits:
    for method in methods:
        arr = pccbox[f"{split}_{method}"]
        box_data.append(flatten_all(arr))
        fold_stats_all.append(fold_stats(arr))

# === Step 4: Define colors and positions ===
method_colors = ["#1f77b4", "#2ca02c", "#ff7f0e", "#9467bd"]
colors = method_colors * len(splits)

positions = []
start = 1
inner_gap = 0.6
group_gap = 2

for i in range(len(splits)):
    group_pos = [start + inner_gap * j for j in range(len(methods))]
    positions.extend(group_pos)
    start = group_pos[-1] + group_gap

print("Boxplot positions:", positions)

# === Step 5: Plotting ===
plt.figure(figsize=(14, 7))
jitter = 0.15

# Plot fold-wise min-max lines first (under boxplots)
for i, stats_list in enumerate(fold_stats_all):
    x_pos = positions[i]
    for stat in stats_list:
        plt.vlines(x_pos + random.uniform(-jitter, jitter), stat['min'], stat['max'], color='gray', alpha=0.3, linewidth=1, zorder=5)

# Plot boxplots
box = plt.boxplot(box_data, patch_artist=True, positions=positions, widths=0.6)
for patch, color in zip(box["boxes"], colors):
    patch.set_facecolor(color)
for whisker in box["whiskers"]:
    whisker.set(color="black", linestyle="--")
for flier in box["fliers"]:
    flier.set(marker="o", color="black", alpha=0.4)

# Plot fold medians on top
for i, stats_list in enumerate(fold_stats_all):
    x_pos = positions[i]
    for stat in stats_list:
        plt.plot(x_pos + random.uniform(-jitter, jitter), stat['median'],
                 'ko', alpha=0.8, markersize=6, zorder=10)

# === Step 6: Formatting ===
group_centers = []
for i in range(len(splits)):
    group = positions[i*len(methods):(i+1)*len(methods)]
    group_centers.append(np.mean(group))

plt.xticks(group_centers, ["Training", "Validation", "Test"], fontsize=12)
plt.ylabel("PCC", fontsize=13)
plt.title(f"PCC Distribution Across All Folds and Hyperparameters\nWith SHIPS+ data for ({target})", fontsize=15)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Legend including methods and fold stats
legend_patches = [
    mpatches.Patch(color=method_colors[0], label="No Feature Selection"),
    mpatches.Patch(color=method_colors[1], label="Causal FS"),
    mpatches.Patch(color=method_colors[2], label="Correlation FS"),
    mpatches.Patch(color=method_colors[3], label="Random Forest FS"),
]

min_max_line = mlines.Line2D([], [], color='gray', linewidth=1, label='Fold Min-Max Range')
median_dot = mlines.Line2D([], [], color='black', marker='o', linestyle='None', markersize=6, label='Fold Median')

plt.legend(handles=legend_patches + [min_max_line, median_dot], loc="upper left")

plt.tight_layout()
plt.savefig(f'./figures/pcc_with_fold_shipsplus_noassum_{target}.png', dpi=300)
plt.show()

```
